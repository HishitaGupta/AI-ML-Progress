# 🔹 PYTHON – Theoretical Interview Questions + Answers

### 1. What makes Python different from other languages?

Python is:

* Interpreted (no compilation step)
* Dynamically typed - Python is dynamically typed, meaning variable types are determined at runtime. You don’t declare types explicitly, and a variable can hold values of different types at different points in the program.
* High-level
* Easy syntax
* Large ecosystem (AI/ML, Web, Automation)

---

### 2. Difference between list and tuple?

| List             | Tuple            |
| ---------------- | ---------------- |
| Mutable          | Immutable        |
| Slower           | Faster           |
| Uses more memory | Uses less memory |
| Uses []          | Uses ()          |

Tuple is used when data should not change.

---

### 3. What is a generator?

A generator is a special type of function that **returns one value at a time** using the `yield` keyword instead of returning all values at once.
It saves memory and is efficient for large data.

Example:

```python
def nums():
    for i in range(10):
        yield i
```

---

### 4. What are decorators?

Decorators are functions that **modify the behavior of another function without changing its code**.

Used for:

* Authentication
* Logging
* Timing
* Validation

Example:

```python
def my_decorator(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper
```

---

### 5. Shallow vs Deep copy?

| Shallow Copy     | Deep Copy            |
| ---------------- | -------------------- |
| Copies reference | Copies entire object |
| Changes reflect  | No changes reflect   |
| Uses `copy()`    | Uses `deepcopy()`    |

---

### 6. What is GIL in Python?

GIL (Global Interpreter Lock) allows only **one thread to execute Python bytecode at a time.**
This is why multithreading is less effective in CPU-bound tasks, but works well in I/O-bound tasks.

---

### 7. Difference between multithreading & multiprocessing?

| Multithreading         | Multiprocessing  |
| ---------------------- | ---------------- |
| Same memory            | Separate memory  |
| Faster for I/O         | Best for CPU     |
| Affected by GIL        | Not affected     |
| Low actual parallelism | True parallelism |

---

### 8. What is memory management in Python?

Python does:
✅ Automatic memory allocation
✅ Garbage collection
✅ Reference counting

Unused objects are deleted automatically.

## 🎯 Interview Answer

### 1. **Global Interpreter Lock (GIL)**
- **Concept**: In CPython, the GIL is a mutex that ensures only one thread executes Python bytecode at a time.  
- **Importance**: It simplifies memory management but prevents true parallel execution of threads on multi-core CPUs.  
- **Impact**: Threads are useful for I/O-bound tasks but inefficient for CPU-bound tasks because of the GIL.

---

### 2. **Multithreading**
- **Concept**: Multiple threads run within the same process, sharing memory space.  
- **Importance**: Best for **I/O-bound tasks** (network requests, file I/O) where threads spend time waiting.  
- **Limitation**: Due to the GIL, only one thread runs Python code at a time, so CPU-heavy tasks don’t benefit.  
- **Example**:
  ```python
  import threading

  def task():
      print("Thread running")

  t1 = threading.Thread(target=task)
  t2 = threading.Thread(target=task)
  t1.start(); t2.start()
  t1.join(); t2.join()
  ```
  Output:  
  ```
  Thread running
  Thread running
  ```

---

### 3. **Multiprocessing**
- **Concept**: Multiple processes run independently, each with its own Python interpreter and memory space.  
- **Importance**: Bypasses the GIL, enabling **true parallelism** across multiple CPU cores. Ideal for **CPU-bound tasks** like heavy computations.  
- **Example**:
  ```python
  from multiprocessing import Process

  def task():
      print("Process running")

  p1 = Process(target=task)
  p2 = Process(target=task)
  p1.start(); p2.start()
  p1.join(); p2.join()
  ```
  Output (order may vary):  
  ```
  Process running
  Process running
  ```

---

### ✅ Final Summary
- **GIL**: Only one thread executes Python code at a time.  
- **Multithreading**: Good for I/O-bound tasks, limited by GIL.  
- **Multiprocessing**: Good for CPU-bound tasks, bypasses GIL with true parallelism.
