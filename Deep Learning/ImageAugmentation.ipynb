
Imbalanced Dataset - overfitted model - more bias towards one class than other - solution - balance out data - dont use accuracy to check such data - use confusin matrix or classification report

upsampling - creating synthetic data points to balance out data
downsampling

Augmentation 
- in laymans term , creating more data from given data
- can be done for voice, image data
- voice data - 1 d data - amplitude vs time
            - Augmentation techniques- (Pitch Shift, Adding Noise, Removing Empty Speech, Increasing/ Decreasing DB) 

- image augmentation - techniques - tilting the image, Adding noise, Blurring , Zooming IN, Zooming Out, Resizing

- libraries like augmentor , data image augmentor can be used to automate the augmentation process


```python
import numpy as np
import pandas as pd
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from skimage import exposure
from skimage.util import random_noise
from skimage import transform
from cv2 import resize
import random
```

```python
img = mpimg.imread('cat.jpg')
plt.imshow(img)
```

```python
# RESCALING

img_rescale = resize(img,(150,150))
plt.imshow(img_rescale)
```

```python
# FLIPPING

horizontal_flip = np.fliplr(img)
plt.imshow(horizontal_flip)
```

```python
# ROTATION

from skimage import transform
trans_img = transform.rotate(img, random.uniform(-40,40))
plt.imshow(trans_img)
```

```python
# Adding Noise

img_noise = random_noise(img, mode ='s&p', clip=True)
plt.imshow(img_noise)
```
