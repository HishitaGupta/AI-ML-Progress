```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Python/Churn_Modelling.csv')

df.head()
```

```python

X = df.iloc[:, 3:13].values
y= df.iloc[:,13].values
```

```python
df.info()
```

```python
# Encoding categorical data

from sklearn.preprocessing import OneHotEncoder, LabelEncoder

labelencoder_X_1 = LabelEncoder()
X[:,1] = labelencoder_X_1.fit_transform(X[:,1])

labelencoder_X_2 = LabelEncoder()
X[:,2] = labelencoder_X_2.fit_transform(X[:,2])

X
```

```python
# Splitting the dataset to training and test set
from sklearn.model_selection import train_test_split

X_train,X_test,y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

##### Classical Decision tree model

```python
# Classical ML
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,pred)

print(cm)
```

```python
from sklearn.metrics import accuracy_score

accuracyScore = accuracy_score(y_test, pred)

print(accuracyScore)
```

##### ANN

```python
# Importing keras libraries and packages

import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialsing the ANN
classifier = Sequential()

# units = Neurons, It's an art, comes by experience, input+output/2
# Initialiser  = how your weights are updated
# relu activation function
# input_dim = input shape( shape of the data, how many features do we have in data)

# Adding the input layer and first hidden layer
classifier.add(Dense(units=6, kernel_initializer ='uniform', activation='relu', input_dim=10))

# Adding the second hidden layer
classifier.add(Dense(units=6, kernel_initializer ='uniform', activation='relu'))

# Adding the third hidden layer
classifier.add(Dense(units=6, kernel_initializer ='uniform', activation='relu'))

# Adding the output layer
classifier.add(Dense(units=1, kernel_initializer ='uniform', activation='sigmoid'))




```

```python
# Compiling the ANN

classifier.compile(optimizer ='adam', loss='binary_crossentropy', metrics =['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size=200, epochs= 20)
```

```python
# Making predictions and evaluating the model
# Predicting the test set results

y_pred_Ann = classifier.predict(X_test)
y_pred_Ann = (y_pred_Ann > 0.5)

cm2 = confusion_matrix(y_test, y_pred_Ann)
print(cm2)
```

```python
# Accuracy Score
from sklearn.metrics import accuracy_score

score =accuracy_score(y_test, y_pred_Ann)
print(score*100)
```

So, as we have already seen, by using 2 hidden layered ANN, we aren't able to get beyond 80% accuracy, ofcourse we can get higher accuracy by increasing the epochs & doing some hyperparameter optimisation, so on the other hand, classical ML models can be tuned further.

What is K-Fold Cross Validation?

It’s a technique to evaluate how well your model generalizes —
meaning: how good your model will perform on unseen data (not just the data it was trained on).

Instead of doing one single train-test split, K-Fold CV splits your dataset into K parts (folds) and runs training + testing K times — each time using a different fold as the test set.

```python
#Importing few libraries for k-folds cross validation on a Keras deep learning model using scikit-learn’s wrapper.

from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense

#Defining the neural network
def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 10))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier

# Wrapping it into a scikit-learn compatible classifier
classifier = KerasClassifier(model = build_classifier, batch_size = 10, epochs = 10)

# So now classifier behaves like a normal scikit-learn model (with .fit() and .predict() methods).

# Applying K-Fold Cross Validation
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv =10, n_jobs = -1 )
mean1 = accuracies.mean()
variance = accuracies.std()
```

cross validation / cv - runs the model with different train test split to identify the best split

##### Hyper Parameter Optimisation

```python
#Improvising the ANN - Hyper parameter optimization

from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense


def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 10))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier


classifier = KerasClassifier(model = build_classifier)

parameters = {'batch_size': [10, 20],
              'epochs': [5, 6]}

              # Total combinations: 4

grid_search = GridSearchCV(estimator = classifier, param_grid = parameters, scoring = 'accuracy', n_jobs = -1, cv = 5)
grid_search = grid_search.fit(X_train, y_train)

grid_search.best_params_
grid_search.best_score_
```

##### Improvising the performance of ANN
- Increasing the number of hidden layers
- Increasing the number of epochs
- Fine-tuning the parameters - optimizers, batch_size, epochs
- Dropout regularization: Dropout(0.25)
Overfitting problem --> training accuracy increases, but test accuracy is low
