```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv('AirPassengers.csv')
print(df.head())
```

```python
df.info()
```

```python
df['Month'] = pd.to_datetime(df['Month'])
```

```python
# 2️⃣ Set 'Month' as the DataFrame index
df.set_index('Month', inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
# Perform seasonal decompose

result = seasonal_decompose(df, model='multiplicative')
```

```python
# PLot the components

plt.rcParams.update({'figure.figsize':(10,10)})
result.plot()
plt.show()
```

```python
# Perform Additive Decomposition

result = seasonal_decompose(df, model='additve')
```

```python
# PLot the components

plt.rcParams.update({'figure.figsize':(10,10)})
result.plot()
plt.show()
```

```python
# Residual graph differs
# The magnitude of seasonal variation in data appears to increase with the increasing trend
# Therefore multiplicative decomposition might  be appropriate
```
