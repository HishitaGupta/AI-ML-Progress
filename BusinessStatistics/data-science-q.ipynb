‚úÖ **Answer:**

## 6. Explain **Bayes' Theorem**

‚úÖ **Answer:**

> Bayes' theorem allows us to update our belief about an event based on new evidence. It combines prior probability and likelihood to give posterior probability.

[
P(A|B) = \frac{P(B|A)\cdot P(A)}{P(B)}
]

**Pro Tip:**
Mention: *"Naive Bayes is built on this."*

# ‚úÖ Q15. **How do you design an A/B test?** (DETAILED VERSION)

## 1. What is A/B Testing (in simple words)?

A/B testing is an **experiment** where you compare:

* Version A ‚Üí original (control)
* Version B ‚Üí new version (variant)

You randomly split your users and show each group a different version.
Then you measure **which one performs better** using statistics.

It is basically:
**Science + Data + Business decision making**

---

## 2. Why companies use A/B testing (REAL BUSINESS REASON)

Used to improve:

* Conversion rate
* Revenue
* Clicks
* Engagement
* Retention

For example:

| Change           | Goal            |
| ---------------- | --------------- |
| New button color | More clicks     |
| New pricing page | Higher sales    |
| New thumbnail    | More watch time |
| Different ad     | More purchases  |

Big companies like Google, Meta, Netflix, Amazon run **hundreds of A/B tests DAILY.**

---

## 3. Interview-Level Step-by-Step Method (Say this in interview üëá)

When interviewer asks:

> How do you design an A/B test?

Your PERFECT answer:

> I follow a systematic approach:
>
> 1. I define a clear business goal (for example: increase conversion rate).
> 2. I identify a primary success metric (like CTR, conversion, revenue).
> 3. I divide users randomly into two groups ‚Äî Control and Variant.
> 4. I apply the change only to the Variant group.
> 5. I define the sample size and test duration to ensure statistical power.
> 6. After the experiment, I perform hypothesis testing to find statistical significance.
> 7. Based on the p-value and confidence interval, I accept or reject the change.

This answer alone is **top 1% level**.

---

## 4. Mathematical Explanation (for clarity)

Let:

* CR‚ÇÅ = Conversion rate of A
* CR‚ÇÇ = Conversion rate of B

We define:

**Null Hypothesis (H‚ÇÄ):**
[
CR_1 = CR_2
]

**Alternative Hypothesis (H‚ÇÅ):**
[
CR_1 \neq CR_2
]

Then compute:

* z-score or t-score
* p-value

If **p ‚â§ 0.05** ‚Üí statistically significant ‚Üí choose winner

---

## 5. How to choose sample size & test duration

Very strong interview point:

You decide size based on:

* Baseline conversion rate
* Expected lift (ex: +5%)
* Power (usually 80%)
* Confidence (usually 95%)

Formula (intuition only):

> The smaller the improvement you want to detect, the larger the sample you need.

Good line to say:

> I calculate sample size before testing to avoid under-powered experiments.

---

## 6. Common Mistakes to avoid (Interview GOLD)

Say at least one of these:

‚úÖ Not stopping test too early
‚úÖ Random split (remove bias)
‚úÖ Same time period for both versions
‚úÖ Same traffic sources
‚úÖ Only change **ONE variable**

If you say this ‚Äî you sound senior.

---

## 7. Python Example (for factual backing)

```python
from scipy.stats import ttest_ind

control = [0,1,0,0,1,0,1,0,0]
variant = [1,1,0,1,1,1,0,1,1]

t_stat, p_value = ttest_ind(control, variant)

if p_value < 0.05:
    print("Statistically Significant")
else:
    print("Not Significant")
```
