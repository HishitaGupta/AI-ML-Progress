```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../EDA/CustomerChurn.csv")
```

```python
df.head()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges,errors='coerce')
```

```python
df.info()
```

```python
df.dropna(how='any',inplace=True)
```

```python
df.Churn.value_counts()/len(df)*100
```

```python
# Churn Rate  = 27%
```

```python
df.columns
```

```python
X = df.drop(['customerID','Churn'],axis=1)
y = df.Churn.values
```

```python
X = pd.get_dummies(
    X,
    columns=['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'],
    drop_first=True,
    dtype=int
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,y_train, y_test = train_test_split(X, y, test_size=0.25, train_size=0.75)
```

```python
(len(y_test)/len(df))*100
```

```python
type(X_test)
```

```python
X_test
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train_sc
```

Decision Tree Clasifier

```python
# Call the DT Classifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

#Initiating the classifier
model_dt = DecisionTreeClassifier()

#Passing the data to classifier
model_dt.fit(X_train_sc,y_train)
```

```python
y_pred_dt = model_dt.predict(X_test_sc)
```

```python
print(accuracy_score(y_test,y_pred_dt)*100)

# Accuracy not correct measure of the model if data is imbalanced
# use classification report
# no good values there due to class imbalance problem
# sol - upsampling , downsampling
```

### Techniques to correct Class Imbalance

#### Oversampling

```python
from imblearn.over_sampling import RandomOverSampler

# Upsample the minority class using RandomOverSampler
oversampler = RandomOverSampler()
X_train_upsampled,y_train_upsampled = oversampler.fit_resample(X_train_sc,y_train)

# Model Building
model_dt2 = DecisionTreeClassifier()
model_dt2.fit(X_train_upsampled,y_train_upsampled)

# Predict
y_pred_dt2 = model_dt2.predict(X_test_sc)

# Accuracy
print(accuracy_score(y_test,y_pred_dt2)*100)

# Classifciation Metrics
dt2_report = classification_report(y_test,y_pred_dt2)
print(dt2_report)
```

##### Check Excel Sheet

#### UnderSampling

```python
from imblearn.under_sampling import RandomUnderSampler

# Upsample the minority class using RandomOverSampler
undersampler = RandomUnderSampler()
X_train_undersampled,y_train_undersampled = undersampler.fit_resample(X_train_sc,y_train)

# Model Building
model_dt2 = DecisionTreeClassifier()
model_dt2.fit(X_train_undersampled,y_train_undersampled)

# Predict
y_pred_dt2 = model_dt2.predict(X_test_sc)

# Accuracy
print(accuracy_score(y_test,y_pred_dt2)*100)

# Classifciation Metrics
dt2_report = classification_report(y_test,y_pred_dt2)
print(dt2_report)
```

#### SMOTEENN

##### Synthetic Minority Oversampling Technique - Edited Nearest Neigbour

##### work by creating synthetic examples for minority class (not copies), take references from class distribution, understand the pattern ie applies SMOT then undersampling technique to remove records

```python
from imblearn.combine import SMOTEENN

# Create a smoteenn project
smote_enn = SMOTEENN(random_state=42)

# Resample the data
X_train_resampled,y_train_resampled = smote_enn.fit_resample(X_train_sc,y_train)

# Model Building
model_dt3 = DecisionTreeClassifier()
model_dt3.fit(X_train_resampled,y_train_resampled)

# Predict
y_pred_dt3 = model_dt3.predict(X_test_sc)

# Accuracy
print(accuracy_score(y_test,y_pred_dt3)*100)

# Classifciation Metrics
dt3_report = classification_report(y_test,y_pred_dt3)
print(dt3_report)
```

#### SMOT

##### work by creating synthetic examples for minority class (not copies), take references from class distribution, understand the pattern

```python
from imblearn.over_sampling import SMOTE

# Create a smoteenn project
smote = SMOTE(random_state=42)

# Resample the data
X_train_resampled,y_train_resampled = smote.fit_resample(X_train_sc,y_train)

# Model Building
model_dt4 = DecisionTreeClassifier()
model_dt4.fit(X_train_resampled,y_train_resampled)

# Predict
y_pred_dt4 = model_dt4.predict(X_test_sc)

# Accuracy
print(accuracy_score(y_test,y_pred_dt4)*100)

# Classifciation Metrics
dt4_report = classification_report(y_test,y_pred_dt4)
print(dt4_report)
```

#### ADASYNN

##### Adaptive Synthetic Sampling Approach for Imbalanced Learning


```python
from imblearn.over_sampling import ADASYN

# Create a smoteenn project
adasyn = ADASYN(random_state=42)

# Resample the data
X_train_resampled,y_train_resampled = adasyn.fit_resample(X_train_sc,y_train)

# Model Building
model_dt5 = DecisionTreeClassifier()
model_dt5.fit(X_train_resampled,y_train_resampled)

# Predict
y_pred_dt5 = model_dt5.predict(X_test_sc)

# Accuracy
print(accuracy_score(y_test,y_pred_dt5)*100)

# Classifciation Metrics
dt5_report = classification_report(y_test,y_pred_dt5)
print(dt5_report)
```

#### AllKNN

```python
from imblearn.under_sampling import AllKNN

# Create a smoteenn project
allknn = AllKNN(sampling_strategy='auto') 

# Resample the data
X_train_resampled,y_train_resampled = allknn.fit_resample(X_train_sc,y_train)

# Model Building
model_dt6 = DecisionTreeClassifier()
model_dt6.fit(X_train_resampled,y_train_resampled)

# Predict
y_pred_dt6 = model_dt6.predict(X_test_sc)

# Accuracy
print(accuracy_score(y_test,y_pred_dt6)*100)

# Classifciation Metrics
dt6_report = classification_report(y_test,y_pred_dt6)
print(dt6_report)
```

#### TOMEKLINKS

```python
from imblearn.under_sampling import TomekLinks
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Apply TomekLinks undersampling
tomek = TomekLinks(sampling_strategy='auto')  
X_train_resampled, y_train_resampled = tomek.fit_resample(X_train_sc, y_train)

# Model Building
model_dt_tomek = DecisionTreeClassifier(random_state=42)
model_dt_tomek.fit(X_train_resampled, y_train_resampled)

# Predict
y_pred_tomek = model_dt_tomek.predict(X_test_sc)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred_tomek) * 100)

# Classification Metrics
tomek_report = classification_report(y_test, y_pred_tomek)
print(tomek_report)
```

#### SMOTTOMEK

```python
from imblearn.combine import SMOTETomek
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Apply SMOTETomek (oversampling + cleaning undersampling)
smote_tomek = SMOTETomek(sampling_strategy='auto', random_state=42)
X_train_resampled, y_train_resampled = smote_tomek.fit_resample(X_train_sc, y_train)

# Model Building
model_dt_smote_tomek = DecisionTreeClassifier(random_state=42)
model_dt_smote_tomek.fit(X_train_resampled, y_train_resampled)

# Predict
y_pred_smote_tomek = model_dt_smote_tomek.predict(X_test_sc)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred_smote_tomek) * 100)

# Classification Metrics
smote_tomek_report = classification_report(y_test, y_pred_smote_tomek)
print(smote_tomek_report)
```
