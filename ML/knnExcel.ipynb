```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_excel("knn.xlsx",sheet_name="Sheet2")

print(df.head())
```

```python
df = df.drop('Gender_M',axis=1)
```

```python
df_de = pd.get_dummies(df,columns=['Gender'], drop_first=True)

print(df_de.head())
```

```python
import seaborn as sns

Age = df_de["Age"]
Gender_M = df_de["Gender_M"]
Game = df_de["Game"]

sns.scatterplot(x=Age,y=Gender_M,hue=Game)
plt.show()


```
