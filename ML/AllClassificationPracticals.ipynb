```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../EDA/CustomerChurn.csv")
```

```python
df.head()
```

```python
df.TotalCharges = pd.to_numeric(df.TotalCharges,errors='coerce')
```

```python
df.info()
```

```python
df.dropna(how='any',inplace=True)
```

```python
df.Churn.value_counts()/len(df)*100
```

```python
# Churn Rate  = 27%
```

```python
df.columns
```

```python
X = df.drop(['customerID','Churn'],axis=1)
y = df.Churn.values
```

```python
X = pd.get_dummies(
    X,
    columns=['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'],
    drop_first=True,
    dtype=int
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,y_train, y_test = train_test_split(X, y, test_size=0.25, train_size=0.75)
```

```python
(len(y_test)/len(df))*100
```

```python
type(X_test)
```

```python
X_test
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train_sc
```

KNN Classifier

```python
# Call the knn Classifier
from sklearn.neighbors import KNeighborsClassifier

#Initiating
model = KNeighborsClassifier()

#Passing the data to classifier
model.fit(X_train_sc,y_train)
```

```python
y_pred = model.predict(X_test_sc)
```

```python
# Classifcation metrics to check how model is performing

from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_pred)*100)
```

Decision Tree Clasifier

```python
# Call the DT Classifier

from sklearn.tree import DecisionTreeClassifier

#Initiating the classifier
model_dt = DecisionTreeClassifier()

#Passing the data to classifier
model_dt.fit(X_train_sc,y_train)
```

```python
y_pred_dt = model_dt.predict(X_test_sc)
```

```python
print(accuracy_score(y_test,y_pred_dt)*100)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

# Initiating

model_rf = RandomForestClassifier(n_estimators=200)
# sometimes accuracy increases by increasing n_estimators that is number of trees in the forest

model_rf.fit(X_train_sc,y_train)
```

```python
y_pred_rf = model_rf.predict(X_test_sc)
```

```python
print(accuracy_score(y_test,y_pred_rf)*100)
```

```python
y_test
```

```python
from sklearn.metrics import classification_report

knn_report = classification_report(y_test,y_pred)
dt_report = classification_report(y_test,y_pred_dt)
rf_report = classification_report(y_test,y_pred_rf)
```

```python
print(knn_report)
```

```python
print(dt_report)
```

# New Classifiers

#### Naive Bayes Classifier

```python
from sklearn.naive_bayes import BernoulliNB

model_nb = BernoulliNB()

model_nb.fit(X_train_sc,y_train)

y_pred_nb = model_nb.predict(X_test_sc)

print(accuracy_score(y_test,y_pred_nb)*100)
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression()

model_lr.fit(X_train_sc,y_train)

y_pred_lr = model_lr.predict(X_test_sc)

print(accuracy_score(y_test,y_pred_lr)*100)
```

#### SVM Classifier

```python
from sklearn.svm import SVC

model_svm = SVC()

model_svm.fit(X_train_sc,y_train)

y_pred_svm = model_svm.predict(X_test_sc)

print(accuracy_score(y_test,y_pred_svm)*100)
```

```python
naive_report = classification_report(y_test,y_pred_nb)
logistic_report = classification_report(y_test,y_pred_lr)
svm_report = classification_report(y_test,y_pred_svm)
```
