```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# from sklearn.cluster import _k_means_lloyd
```

```python
df = pd.read_csv("./mall.csv")
print(df.info())
```

```python
X = df.iloc[:,[3,4]].values
# Annual Income and Spending Score
```

```python
# Using Dendrogram to find optimal number of clusters

import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(X, method ='ward'))

plt.title("Dendrograms")
plt.xlabel("Clusters")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
# Identified clusters = 4
# hc was giving clusters = 5
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters =4, linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
# Visulaising

plt.scatter(X[y_hc==0,0], X[y_hc==0,1], s=100 , c='red', label='Cluster 1')
plt.scatter(X[y_hc==1,0], X[y_hc==1,1], s=100 , c='blue', label='Cluster 2')
plt.scatter(X[y_hc==2,0], X[y_hc==2,1], s=100 , c='green', label='Cluster 3')
plt.scatter(X[y_hc==3,0], X[y_hc==3,1], s=100 , c='yellow', label='Cluster 4')
# plt.scatter(X[y_hc==4,0], X[y_hc==4,1], s=100 , c='cyan', label='Cluster 5')

#centroids
# plt.scatter(hc.cluster_centers_[:,0], hc.cluster_centers_[:,1], s=300, c ='black', label='Centroids')


plt.title("Cluster of Clients")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score (0-100)")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters =5, linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
# Visulaising

plt.scatter(X[y_hc==0,0], X[y_hc==0,1], s=100 , c='red', label='Cluster 1')
plt.scatter(X[y_hc==1,0], X[y_hc==1,1], s=100 , c='blue', label='Cluster 2')
plt.scatter(X[y_hc==2,0], X[y_hc==2,1], s=100 , c='green', label='Cluster 3')
plt.scatter(X[y_hc==3,0], X[y_hc==3,1], s=100 , c='yellow', label='Cluster 4')
plt.scatter(X[y_hc==4,0], X[y_hc==4,1], s=100 , c='cyan', label='Cluster 5')

#centroids
# plt.scatter(hc.cluster_centers_[:,0], hc.cluster_centers_[:,1], s=300, c ='black', label='Centroids')


plt.title("Cluster of Clients")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score (0-100)")
plt.legend()
plt.show()
```
