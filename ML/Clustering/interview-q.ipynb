

### 1. What is clustering?

**Perfect answer:**

> Clustering is an unsupervised learning technique used to group similar data points into clusters based on feature similarity, without using any labeled output.

---

### 2. Difference between classification and clustering?

| Classification                  | Clustering              |
| ------------------------------- | ----------------------- |
| Supervised                      | Unsupervised            |
| Needs labeled data              | No labels               |
| Predicts predefined class       | Discovers hidden groups |
| Accuracy can be measured easily | Harder to evaluate      |

**Answer (say this):**

> Classification uses labeled data to assign predefined categories, while clustering discovers natural groupings in unlabeled data.

---

### 3. Name different clustering algorithms

You must memorize these:

* K-Means
* Hierarchical (Agglomerative, Divisive)
* DBSCAN
* Mean Shift
* Gaussian Mixture Models (GMM)
* Spectral Clustering

**Interview line:**

> The most commonly used clustering algorithms are K-Means, Hierarchical Clustering, DBSCAN, and Gaussian Mixture Models.

---

# ✅ K-MEANS (MOST IMPORTANT)

### 4. How does K-Means work?

Perfect, step-by-step answer:

1. Choose **K** centroids
2. Assign each point to nearest centroid
3. Update centroids (mean of points)
4. Repeat until convergence

Say:

> K-Means tries to minimize the within-cluster sum of squares (WCSS).

---

### 5. What is WCSS?

> It measures the total distance between each point and its cluster centroid. Lower WCSS means better clustering.

---

### 6. How do you choose K?

* Elbow method
* Silhouette score
* Domain knowledge

Say:

> I generally use the elbow method or silhouette score.

---

### 7. Limitations of K-Means (VERY important)

Must say at least 3 of these:

* Need to predefine K
* Sensitive to outliers
* Works poorly for non-spherical data
* Problem with varying densities

---

# ✅ HIERARCHICAL CLUSTERING

### 8. What is hierarchical clustering?

> It builds a hierarchy (tree-like structure) of clusters using either bottom-up (agglomerative) or top-down (divisive) approaches.

**Dendrogram** is the output.

Good for visualization.

---

### 9. Agglomerative vs Divisive

| Agglomerative  | Divisive       |
| -------------- | -------------- |
| Bottom-Up      | Top-Down       |
| Merge clusters | Split clusters |
| More common    | Less common    |

---

# ✅ DBSCAN (VERY IMPORTANT FOR REAL USE)

### 10. What is DBSCAN?

> DBSCAN is a density-based clustering method that groups points that are closely packed and marks points in low-density regions as outliers.

Uses:

* ε (epsilon) – radius
* MinPts – minimum points

---

### 11. DBSCAN vs K-Means

| K-Means                 | DBSCAN           |
| ----------------------- | ---------------- |
| Needs K                 | No K required    |
| Sensitive to outliers   | Handles outliers |
| Only spherical clusters | Any shape        |

Say this:

> If the data contains outliers or arbitrary-shaped clusters, I prefer using DBSCAN.

---



# ✅ MODEL EVALUATION IN CLUSTERING

### 13. How do you evaluate clustering?

Most important answers:

• Silhouette score
• Davies–Bouldin index
• Calinski–Harabasz index

Say:

> Since clustering is unsupervised, we use internal validation metrics such as silhouette score.

---

# ✅ REAL-WORLD QUESTIONS (CRITICAL)

### 14. When do you choose DBSCAN over K-Means?

> When data contains noise and clusters of arbitrary shape, DBSCAN is a better choice.

---

### 15. What if my data is not linearly separable?

> I’d use DBSCAN or Spectral Clustering instead of K-Means.

---

### 16. What if my dataset is very large?

> I’d prefer MiniBatch K-Means for scalability.

---

# ✅ ADVANCED QUESTIONS

### 17. Why scaling is important in clustering?

> Because distance-based algorithms like K-Means are sensitive to scale, so I always apply normalization or standardization.

Very important to say.

---

### 18. Difference between clustering and dimensionality reduction?

| Clustering      | Dimensionality Reduction |
| --------------- | ------------------------ |
| Groups data     | Reduces features         |
| Output: labels  | Output: new features     |
| Example: KMeans | Example: PCA             |
