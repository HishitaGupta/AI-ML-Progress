```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# from sklearn.cluster import _k_means_lloyd
```

```python
df = pd.read_csv("./mall.csv")
print(df.info())
```

```python
X = df.iloc[:,[3,4]].values
# Annual Income and Spending Score
```

```python
#Fitting the mean shift clustering to the mall dataset

from sklearn.cluster import MeanShift

ms = MeanShift(bandwidth=20)
# Adjust the bandwidth parameter to influence number of clusters

ms.fit(X)
labels = ms.labels_
cluster_centers = ms.cluster_centers_


```

```python
labels # 7 clusters
```

```python
cluster_centers
```

```python
#Visulaisding the clusters

n_clusters = len(np.unique(labels))
colors = ['red','blue','green','yellow','cyan','purple','pink']

for i in range(n_clusters):
    cluster = X[labels==i]
    plt.scatter(cluster[:,0], cluster[:,1], s=100, c=colors[i],label=f'Cluster {i+1}')

plt.title("Cluster of Clients")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score (0-100)")
plt.legend()
plt.show()
```
