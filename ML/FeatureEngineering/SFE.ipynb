```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SequentialFeatureSelector 



from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('../../EDA/CustomerChurn.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
df.dropna(how='any',inplace=True)
```

```python
df.drop(['customerID'], axis ='columns',inplace=True)
```

```python
#Dummy Encoding

df = pd.get_dummies(df,drop_first=True)
```

```python
df.info()
```

```python
# Considering Churn -> Y Variable
# Other Columns -> X Variable

X= df.drop(['Churn_Yes'],axis=1)
y = df['Churn_Yes']
```

```python
# Split the data

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=40)
```

```python
# Base Model

model_orig = LogisticRegression()
model_orig.fit(X_train,y_train)
y_pred_orig = model_orig.predict(X_test)

accuracy = accuracy_score(y_test,y_pred_orig)
print(accuracy_score(y_test,y_pred_orig))
```

```python
# Apply Successive Feature Selection to select 5 best features


```

```python
model_sfs = LogisticRegression()
sfs = SequentialFeatureSelector(model_sfs,k_features=5,scoring='accuracy',cv=0)

# Fit SFS on training data

sfs.fit(X_train, y_train)

```

```python
# Get the selected features 

selected_features =list(sfs.k_feature_names_)
X_train_selection = X_train[selected_features]
X_test_selection = X_test[selected_features]
```

```python
selected_features
```

```python
# SFS Based Model


model_sfs.fit(X_train_selection,y_train)
y_pred_sfs = model_sfs.predict(X_test_selection)

accuracy_sfs = accuracy_score(y_test,y_pred_sfs)
print(accuracy_sfs)
```
