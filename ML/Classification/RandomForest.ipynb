```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../EDA/CustomerChurn.csv')
print(df.columns)
```

```python
df.info()
```

```python
df.TotalCharges=pd.to_numeric(df.TotalCharges,errors='coerce')
```

```python
df.info()
```

```python
df.dropna(how='any',inplace=True)
```

```python
(df.Churn.value_counts()/len(df.Churn))*100
```

```python

X = df.drop(['customerID','Churn'],axis=1)
y= df.Churn.values

print(df.columns)
print(type(X))

```

```python
# Feature Encoding

X = pd.get_dummies(X,columns=['gender', 'SeniorCitizen', 'Partner', 'Dependents','PhoneService','MultipleLines', 'InternetService', 'OnlineSecurity','OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV','StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'],drop_first=True)

```

```python
print(X.info())
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
print(len(y_test))
print(len(X_train))
```

```python
# Feature Scaling


from sklearn.preprocessing import StandardScaler

sc= StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)

```

```python
X_train_sc
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initiating

model = RandomForestClassifier()
# sometimes accuracy increases by increasing n_estimators that is number of trees in the forest

model.fit(X_train_sc,y_train)
```

```python

# pass testing data for predictions

y_pred = model.predict(X_test_sc)
```

```python
# Classification Metrics

from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred) 
```
